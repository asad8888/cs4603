```python
def print_token_usage(chat_completion):
    pprint(f"Model used: {chat_completion.model}")
    pprint(f"Input tokens: {chat_completion.usage.prompt_tokens}")
    pprint(f"Output tokens: {chat_completion.usage.completion_tokens}")
    pprint(f"Total tokens: {chat_completion.usage.total_tokens}")
```

```python
print_token_usage(chat_completion)
```
