## 1. Basic PromptTemplate — single variable

```python
template1 = PromptTemplate.from_template(
    "Explain {topic} in exactly 2 concise bullet points."
)
chain1 = template1 | llm | parser
print(chain1.invoke({"topic": "vector embeddings"}))
```

## 2. Multiple input variables

```python
template2 = PromptTemplate(
    input_variables=["audience", "topic", "tone"],
    template="Write a {tone} 4-line explanation of {topic} for {audience}.",
)
chain2 = template2 | llm | parser
print(chain2.invoke({"audience": "beginners", "topic": "RAG", "tone": "friendly"}))
```

## 3. ChatPromptTemplate with system/human roles

```python
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a concise teaching assistant."),
    ("human", "Give 3 tips for learning {topic}."),
])
chain3 = chat_template | llm | parser
print(chain3.invoke({"topic": "LangChain"}))
```

## 4. Partial templates — lock in a variable

```python
base_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}."),
    ("human", "Explain {topic} in one sentence."),
])

# Lock in the role so we only need to pass topic at invocation time
instructor_template = base_template.partial(role="helpful instructor")
chain4 = instructor_template | llm | parser
print(chain4.invoke({"topic": "Databricks Vector Search"}))
```

## 5. Batch invocation

```python
batch_inputs = [
    {"topic": "embeddings"},
    {"topic": "chunking"},
    {"topic": "similarity search"},
]
results = chain3.batch(batch_inputs)
for inp, result in zip(batch_inputs, results):
    print(f"\n--- {inp['topic']} ---\n{result}")
```
